```python
import pandas as pd

#loading the dataset
df = pd.read_csv('data/properties.csv')

# dimension check
print("\n ---Data Info ---")
print(df.info())

# statistical summary
print("\n --- Statistical Summary ---")
print(df.describe())
```

```python

# checking for missing values
print(df.isnull().sum())
```

```python
from src.dataset_filter import is_chon_only

# Filter for CHON only
print(f"Original count: {len(df)}")
df['is_chon'] = df['Molecule'].apply(is_chon_only)
df_clean = df[df['is_chon'] == True].copy()
```

```python
# Filter for reasonable size (e.g., Heavy Atoms < 100)
df_clean = df_clean[df_clean['Heavy Atoms Count'] < 100]
```

```python
# Deduplicate
df_clean = df_clean.drop_duplicates(subset='Molecule')
```

```python
# Save for the pipeline
df_clean[['Molecule', 'Molecule name']].to_csv('data/cleaned_wiki_chon.csv', index=False)
```
